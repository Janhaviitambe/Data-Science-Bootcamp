```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD')
df
```

```python
# Convert 'hour_beginning' to datetime format
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'])
```

```python
# Create a new column for the day of the week
df['day_of_week'] = df['hour_beginning'].dt.day_name()
```

```python
# Filter data for only weekdays
weekday_data = df[df['day_of_week'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])]
```

```python
# Group by day of the week, calculate the average pedestrian count
weekday_counts = weekday_data.groupby('day_of_week')['Pedestrians'].mean().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
)
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
weekday_counts.plot(kind='line', marker='o')
plt.title("Average Pedestrian Counts by Weekday")
plt.xlabel("Day of the Week")
plt.ylabel("Average Pedestrian Count")
plt.grid(True)
plt.show()
```
