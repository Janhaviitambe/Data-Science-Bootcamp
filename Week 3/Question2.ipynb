```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('https://data.cityofnewyork.us/api/views/6fi9-q3ta/rows.csv?accessType=DOWNLOAD')
df
```

```python
# Convert 'hour_beginning' to datetime format
df['hour_beginning'] = pd.to_datetime(df['hour_beginning'])
```

```python
# Filter data for the year 2019 and Brooklyn Bridge location
df_2019 = df[(df['hour_beginning'].dt.year == 2019) & (df['location'] == 'Brooklyn Bridge')]
```

```python
# One-hot encode 'weather_summary' to create separate columns for each weather condition
df_2019_encoded = pd.get_dummies(df_2019, columns=['weather_summary'])

# Select only numeric columns to avoid errors
df_2019_encoded_numeric = df_2019_encoded.select_dtypes(include=['number'])
```

```python
# Calculate correlation matrix on the filtered numeric columns
correlation_matrix = df_2019_encoded_numeric.corr()

# Show correlations of 'Pedestrians' with weather-related columns
weather_correlations = correlation_matrix['Pedestrians'].filter(like='weather_summary')
weather_correlations
```
